# This is a notebook for calculating the collection efficiency of a RRDE

```python
import numpy as np
import math as m
```

```python
#change these parameters
r1 = 4 #mm
r2 = 5 #mm
r3 = 7 #mm
```

```python
#useful definitions
a = (r2/r1)**3 -1
b = ((r3**3)/(r1**3) - (r2**3)/(r1**3))

c = a/b
d = c*(1+a+b)
```

\begin{equation*}
N = 1-F\biggl(\frac{\alpha}{\beta}\biggr) + \beta^{\frac{2}{3}}\biggl(1-F(\alpha)\biggr) - \biggl(1+\alpha+\beta\biggr)^{\frac{2}{3}}\Biggl(1-F\biggl(\bigl(\frac{\alpha}{\beta}\bigr)(1+\alpha+\beta)\biggr)\Biggr)
\end{equation*}

\begin{equation*}
F(\theta) = \biggl(\frac{\sqrt{3}}{4\pi}\biggr)ln\Biggl[\frac{\bigl(1+\theta^{\frac{1}{3}}\bigr)^3}{1+\theta}\Biggr] + \biggl(\frac{3}{2\pi}\biggr)tan^{-1}\Biggl(\frac{2\theta^\frac{1}{3}-1}{3^{\frac{1}{2}}}\Biggr)+\frac{1}{4}
\end{equation*}


```python
#defining functions
def f(theta):
    func = (m.sqrt(3)/(4*m.pi))*np.log(((1+theta**(1/3))**3)/(1+theta)) + (3/(2*m.pi))*m.atan((2*theta**(1/3) - 1)/(3**0.5))+0.25
    return func

#test func
#Fab = (m.sqrt(3)/(4*m.pi))*np.log(((1+c**(1/3))**3)/(1+c)) + (3/(2*m.pi))*m.atan((2*c**(1/3) - 1)/(3**0.5))+0.25
```

```python
#calculation collection efficiency
N = 1 - f(c) + b**(2/3)*(1-f(a)) - (1+a+b)**(2/3) * (1-f(d))
print('the collection efficiency is:', round(N*100,1), '%')
```
